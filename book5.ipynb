# custom dataset traning 

```python
import requests
import zipfile
from pathlib import Path

```

```python
data_path = Path("dataset")
image_path = data_path / "images"

if image_path.is_dir():
    print("Image directory already exists.")
else:
    print("Creating image directory...")
    image_path.mkdir(parents=True, exist_ok=True)
    with open(data_path / "images.zip", "wb") as f:
        print("Downloading images...")
        response = requests.get(
            "https://github.com/mrdbourke/pytorch-deep-learning/raw/refs/heads/main/data/pizza_steak_sushi.zip")
        f.write(response.content)
    with zipfile.ZipFile(data_path/"images.zip", "r") as zip_ref:
        print("Extracting images...")
        zip_ref.extractall(image_path)
```

```python
import torch 
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor
from torch import nn
from PIL import Image

```

```python
test_dataset = datasets.ImageFolder(
    root=image_path/"test",
    transform=ToTensor()
)
train_dataset = datasets.ImageFolder(
    root=image_path/"train",
    transform=ToTensor()
)
```
