## reshaping, stacking, squeezing and unsqueezing

* **reshape**: reshapes the tensor to the given shape
* **stack**: stacks the tensors along a new dimension like vstack and hstack
* **squeeze**: removes single-dimensional entries from the shape
* **unsqueeze**: adds a single-dimensional entry to the shape
* **view**: returns a view of the input with a specific shape
* **permute**: returns a view of the input with a specific permutation of dimensions

```python
y = tf.arange(1., 13.)
y.shape,y
```

while reshaping the tensors should be compatible

```python
y.reshape(4, 3)
```

```python
y.view(3,4)
```

changing view changes the orignal tensor as they share the same memory


```python
y_stacked = tf.stack([y,y,y,y], dim=0)
y_stacked
```

squeeze removes dimensions of size 1

```python
x = tf.squeeze(y)
x.ndim
```

unsqueeze adds a new dimension to the tensor at the dimention mentioned

```python
tf.unsqueeze(x,dim=0).ndim
```
